```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Assign colors based on quadrants
colors = []
for i in range(len(x)):
    if x[i] >= 0 and y[i] >= 0:
        colors.append('red')  # Top-right quadrant
    elif x[i] < 0 and y[i] >= 0:
        colors.append('blue')  # Top-left quadrant
    elif x[i] < 0 and y[i] < 0:
        colors.append('green')  # Bottom-left quadrant
    else:
        colors.append('orange')  # Bottom-right quadrant

# Create scatter plot
plt.scatter(x, y, c=colors)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Quadrant Scatter Plot')

# Show the plot
plt.show()
```

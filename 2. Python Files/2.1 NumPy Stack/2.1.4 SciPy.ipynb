<h2 style = 'color:lightblue'> 2.1.4.1 Introduction to SciPy</h2>

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm

from PIL import Image

from scipy.signal import convolve2d
```

<h2 style = 'color:lightblue'> 2.1.4.2 PDF and CDF</h2>

```python
# Calculate PDF of a normal distribution
x = np.linspace(-5, 5, 1000)
fx = norm.pdf(x, loc=0, scale=1)
plt.plot(x, fx); # First Argument is x-axis, second is y-axis
```

```python
# Calculate CDF of a normal distribution
Fx = norm.cdf(x, loc=0, scale=1)
plt.plot(x, Fx);
```

```python
# Calculate Log PDF of a normal distribution
logfx = norm.logpdf(x, loc=0, scale=1)
plt.plot(x, logfx);
```

<h2 style = 'color:lightblue'> 2.1.4.3 Convolution</h2>

Convolution is a very common technique in deep learning, signal processing, computer vision, and machine learning. A 2D convolution is a convolution that operates on a 2D input signal and produces a 2D, and hence we need to remove the color information from the input signal.

```python
im = Image.open('../../0. Images/lena.png')
gray = np.mean(im, axis=2)
plt.imshow(gray, cmap='gray')
```

```python
# Create a 2D gaussian filter
x = np.linspace(-6, 6, 50)
fx = norm.pdf(x, loc=0, scale=1)
```

```python
Image.open('../../0. Images/Gaussian Filter.png')
```

```python
filt = np.outer(fx, fx)
```

```python
plt.imshow(filt, cmap='gray')
```

```python
out = convolve2d(gray, filt)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')

plt.subplot(1, 2, 2)
plt.imshow(out, cmap='gray')
```
